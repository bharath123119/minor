```python
import pandas as pd  
import matplotlib.pyplot as plt
df=pd.read_csv("C:/Users/Bharath/OneDrive/Documents/Downloads/supply_chain_data.csv")
df
```

# Plot the Product type value_counts

```python
value_counts=df['Product type'].value_counts()
value_counts
```

```python
value_counts.plot(kind='bar')
plt.xlabel('Product Type')
plt.ylabel('Count')
plt.title('Product Type Distribution')
plt.show()  
```

# Total Revenue Generated by Each Product

```python
Revenue_generated=df.groupby('Product type')['Revenue generated'].sum()
Revenue_generated
```

# plot the pie chart for the Revenue_generated

```python
Revenue_generated.plot(kind='pie', color='orange')
plt.title('Revenue Generated by Product Type')
plt.ylabel('')
plt.show()
```

# Total Availbility In Each Location

```python
Availability=df.groupby('Location')['Availability'].sum()
Availability
```

# plot the Sum of Availability in Each location

```python
Availability.plot(kind='barh',color='green')
plt.xlabel('Availability')
plt.ylabel('Location')  
plt.title('Availability by Location')    
plt.show()  
```

# Gender wise Product Sold

```python
Sum_of_Products_gender=df.groupby('Customer demographics')['Number of products sold'].sum()
Sum_of_Products_gender
```

```python
Sum_of_Products_gender.plot(kind='bar', color='purple')
plt.xlabel('Customer Demographics')     
plt.ylabel('Number of Products Sold')
plt.title('Number of Products Sold by Customer Demographics')
plt.show()
```
